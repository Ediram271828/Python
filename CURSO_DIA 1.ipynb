## Operaciones Matemáticas


```python
# 4to nivel de importancia
# Suma Ordinaria
print("----------------------------")
print("Suma Ordinaria: ")
print(3 + 4)
variable1,variable2 = 20,40 #60
print(variable1 + variable2)

# Resta Ordinaria
print("----------------------------")
print("Resta Ordinaria: ")
print(2- 5)
# Declaración de variables
v1 = 10.4
v2 = 15
resultado = v2- v1
print(resultado)

# 3er nivel de impportancia
# Multiplicación Ordinaria
print("----------------------------")
print("Multiplicación Ordinaria: ")
m1,m2 = 2,3.14
print(m1 * m2) # 6.28
# División Ordinaria
print("División Ordinaria: ")
M1,M2,M3,M4,M5,M6 = 2,5,6,5,4,5
print(M1 / M2)
print(M3 / M4)
print(M5 / M6)

# División de piso
print("----------------------------")
print("División de piso")
print(M1 // M2)
print(M3 // M4)
print(M5 // M6)

print( 6 / 2 * 3)
```

```python
# Potencias
print("Potencias: ")
print(2 ** 3) # 8
P1,P2 = 7,2
print(P1 ** P2)
```

```python
# Area de un circulo
pi = 3.14
r = 4
A = pi * (r ** 2)
print(A)
```

## Operaciones Lógicas

```python
# Empezamos con los operadores lógicos <,>, <=,>=, and, or
b1 = 4>5
b2 = 7<1
b3 = 14<=14
print(b1)
print(b2)
print(b3)
```

```python
# Quieres un sandwich con todo
pan = True
queso = True
mayonesa = True
jamon = False
print(pan & queso & mayonesa & jamon)
print(pan & queso & mayonesa)
print(queso & mayonesa, (mayonesa | jamon))

```
